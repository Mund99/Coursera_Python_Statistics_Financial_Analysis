# Demo a Simple Trading Strategy 
Day trade when MA10 (short signal) > MA50 (long signal)

```python
# Import libraries 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline

# Obtain data 
META = pd.read_csv('data/facebook.csv')

# Set Date as datetime and as index
META['Date'] = pd.to_datetime(META['Date'])
META.set_index('Date', inplace=True)

META.head()
```

```python
# New column - MA10 & MA50 
META['MA10'] = META['Close'].rolling(10).mean()
META['MA50'] = META['Close'].rolling(50).mean()

# Remove empty line 
META = META.dropna()

META.head()
```

```python
# Plot the MA with the close price 
plt.figure(figsize=(15,6))
plt.plot(META.index, META['Close'], label='Close')
plt.plot(META.index, META['MA10'], label='MA10')
plt.plot(META.index, META['MA50'], label='MA50')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('FB Price Data')

plt.legend()

plt.show()
```

```python
# New column - TradeAction 
# TradeAction = 1 if there MA10 > MA50, meaning long one share for that day 
# else TradeAction = 0, meaning no action 

META['TradeAction'] = [1 if META['MA10'].loc[i] > META['MA50'].loc[i] else 0 for i in META.index]

META.head() 
```

```python
# New column - Profit
# Caclulate profit by taking "close price tomorrow" - "close price today" 

META['Close1'] = META['Close'].shift(-1)
META['Profit'] = [META['Close1'].loc[i] - META['Close'].loc[i] if META['TradeAction'].loc[i] == 1 else 0 for i in META.index]

META.head()
```

```python
# Plot Daily Profit & Loss 
plt.figure(figsize=(10,6))

plt.plot(META.index, META['Profit'], label='Daily Profit')
plt.axhline(y=0, color='red', linestyle='dotted')

plt.xlabel('Date')
plt.ylabel('Profit & Loss')
plt.title('FB Trading Daily Profit & Loss')

plt.show() 
```

```python
# New column - CumulativeProfit
# Calculate the cumulative profit day by day using cumsum() 

META['CumulativeProfit'] = META['Profit'].cumsum()

META.head()
```

```python
print(f"Total money you win = {META['CumulativeProfit'].iloc[-2].round(2)}")
```

```python
# Plot Cumulative Profit 

plt.figure(figsize=(10,6))

plt.plot(META.index, META['CumulativeProfit'], label='Cumulative Profit')
plt.axhline(y=0, linestyle='--', color='red')

plt.xlabel('Date')
plt.ylabel('Profit ($)')
plt.title('FB Trading Cumulative Profit')

plt.show() 
```
