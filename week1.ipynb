```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Obtain Data

```python
fb = pd.read_csv('data/facebook.csv')
ms = pd.read_csv('data/microsoft.csv')

fb.head(), ms.head()
```

```python
# Set Date to Pandas.DateTime 
fb['Date'] = pd.to_datetime(fb['Date']) 
ms['Date'] = pd.to_datetime(ms['Date']) 

# Set 'Date' column as the index
fb.set_index('Date', inplace=True)
ms.set_index('Date', inplace=True)

fb.head(), ms.head()
```

```python
# Show the size of the DF 
fb.shape, ms.shape
```

```python
# Statistical summary of the DF 
fb.describe(), ms.describe() 
```

# Slicing using iloc & loc
loc = [row_label,column_label]

iloc = [row_position, column_position]

```python
fb.head()
```

```python
# Select all the price information of FB for year 2015 

# Using loc
fb_2015 = fb.loc['2015-01-01':'2015-12-31']

# Using iloc
fb_2016 = fb.iloc[(fb.index >= '2015-01-01') & (fb.index <= '2015-12-31')] 

fb_2015
```

```python
# Select all the price information of MSFT for year 2016 

# Using loc
ms_2016 = ms.loc['2016-01-01':'2016-12-31']

# Using iloc
ms_2016 = ms.iloc[(ms.index >= '2016-01-01') & (ms.index <= '2016-12-31')] 

ms_2016
```

```python
# Print the opening price of FB on the first row 
fb.iloc[0]['Open']
fb.loc[fb.index[0]]['Open']
```

```python
# Print the closing price of MSFT on the last row 
ms.iloc[-1]['Close']
ms.loc[ms.index[-1]]['Close']
```

# Plot the stock using plot() method

```python
plt.figure(figsize=(10,8))
fb['Close'].plot()
plt.show()
```

```python
# Create a Figure and an array of subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

# Plot the 'Close' data in the first subplot
axes[0].plot(fb['Close'])
axes[0].set_title('Facebook Close Price')

# Plot another data or customization in the second subplot
axes[1].plot(ms['Close'])
axes[1].set_title('Microsoft Close Price')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plots
plt.show()
```
