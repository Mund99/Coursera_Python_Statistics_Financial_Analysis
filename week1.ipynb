# Demo a Simple Trading Strategy 
Day trade when MA10 (short signal) > MA50 (long signal)

```python
# Import libraries 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline

# Obtain data 
FB = pd.read_csv('data/facebook.csv')

# Set Date as datetime and as index
FB['Date'] = pd.to_datetime(FB['Date'])
FB.set_index('Date', inplace=True)

FB.head()
```

```python
# New column - MA10 & MA50 
FB['MA10'] = FB['Close'].rolling(10).mean()
FB['MA50'] = FB['Close'].rolling(50).mean()

# Remove empty line 
FB = FB.dropna()

FB.head()
```

```python
# Plot the MA with the close price 
plt.figure(figsize=(15,6))
plt.plot(FB.index, FB['Close'], label='Close')
plt.plot(FB.index, FB['MA10'], label='MA10')
plt.plot(FB.index, FB['MA50'], label='MA50')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('FB Price Data')

plt.legend()

plt.show()
```

```python
# New column - TradeAction 
# TradeAction = 1 if there MA10 > MA50, meaning long one share for that day 
# else TradeAction = 0, meaning no action 

FB['TradeAction'] = [1 if FB['MA10'].loc[i] > FB['MA50'].loc[i] else 0 for i in FB.index]

FB.head() 
```

```python
# New column - Profit
# Caclulate profit by taking "close price tomorrow" - "close price today" 

FB['Close1'] = FB['Close'].shift(-1)
FB['Profit'] = [FB['Close1'].loc[i] - FB['Close'].loc[i] if FB['TradeAction'].loc[i] == 1 else 0 for i in FB.index]

FB.head()
```

```python
# Plot Daily Profit & Loss 
plt.figure(figsize=(10,6))

plt.plot(FB.index, FB['Profit'], label='Daily Profit')
plt.axhline(y=0, color='red')

plt.xlabel('Date')
plt.ylabel('Profit & Loss')
plt.title('FB Trading Daily Profit & Loss')

plt.show() 
```

```python
# New column - CumulativeProfit
# Calculate the cumulative profit day by day using cumsum() 

FB['CumulativeProfit'] = FB['Profit'].cumsum()

FB.head()
```

```python
# Plot Cumulative Profit 

plt.figure(figsize=(10,6))

plt.plot(FB.index, FB['CumulativeProfit'], label='Cumulative Profit')

plt.xlabel('Date')
plt.ylabel('Profit ($)')
plt.title('FB Trading Cumulative Profit')

plt.show() 
```
